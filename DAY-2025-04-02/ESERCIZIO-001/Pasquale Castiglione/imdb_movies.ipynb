```python
import pandas as pd
import re
```

## Leggere il contenuto del file

```python
df = pd.read_csv("imdb_movies_2024.csv")
df.head()
```

## Ottenere la lista dei film ordinata per titolo

```python
df_sorted_titles = df.copy()

# Estrae il titolo
df_sorted_titles['Title'] = df['Title'].apply(lambda x: x.split(".")[1])

# Ordina in base al titolo
df_sorted_titles.sort_values(by=['Title'])
```

## Sapere quanti film hanno una durata >= 2h

```python
df_duration= df.copy()

# Estrae il numero di ore dalla durata
df_duration['Hour Duration'] = df_duration['Duration'].apply(lambda x: re.search(r'(\d+)h', str(x)).group(1) if re.search(r'(\d+)h', str(x)) else None)

# Rimuove righe con valori nulli
df_duration.dropna(subset=['Hour Duration'], inplace=True)

# Converte in intero
df_duration['Hour Duration'] = df_duration['Hour Duration'].astype(int)

df_duration[df_duration['Hour Duration'] >= 2 ]
```
